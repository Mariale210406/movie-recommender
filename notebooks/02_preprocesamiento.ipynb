```python
df = pd.read_csv('../data/raw/TMDB_all_movies.csv')
print(f"Filas originales: {df.shape[0]}")
```

```python
df = df[df['status'] == 'Released']
print(f"Filas después de filtrar Released: {df.shape[0]}")
```

```python
df = df[df['vote_count'] >= 50]
print(f"Filas después de filtrar vote_count >= 50: {df.shape[0]}")
```

```python
df = df.drop_duplicates(subset='imdb_id')
print(f"Filas después de eliminar duplicados: {df.shape[0]}")
```

```python
df['overview'] = df['overview'].fillna('')
df['genres'] = df['genres'].fillna('Unknown')
df['director'] = df['director'].fillna('Unknown')
df['production_companies'] = df['production_companies'].fillna('Unknown')
df['title'] = df['title'].fillna('Sin título')

print("Valores faltantes manejados")
print(df.isnull().sum())
```

```python
df = df.dropna(subset=['imdb_id'])
df['release_date'] = df['release_date'].fillna('Unknown')
df['imdb_rating'] = df['imdb_rating'].fillna(df['imdb_rating'].median())
df['imdb_votes'] = df['imdb_votes'].fillna(df['imdb_votes'].median())
df['poster_path'] = df['poster_path'].fillna('')

print("✅ Limpieza final completada")
print(df.isnull().sum())
```

```python
df['release_year'] = pd.to_datetime(df['release_date'], errors='coerce').dt.year
df['release_year'] = df['release_year'].fillna(0).astype(int)

print("Año extraído correctamente")
df[['title', 'release_date', 'release_year']].head()
```

```python
df.to_csv('../data/processed/TMDB_clean.csv', index=False)
print(f"✅ Dataset limpio guardado con {df.shape[0]} películas")
```
